```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# AND Dataset
# -------------------------------
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
], dtype=np.float32)

y = np.array([
    [0],
    [0],
    [0],
    [1]
], dtype=np.float32)

# -------------------------------
# Plot Dataset
# -------------------------------
plt.figure(figsize=(5,5))

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i,0], X[i,1], color='red', s=100, label='Class 0' if i==0 else "")
    else:
        plt.scatter(X[i,0], X[i,1], color='blue', s=100, label='Class 1')

plt.title("AND Dataset")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
plt.legend()
plt.show()

# -------------------------------
# Single Layer Perceptron
# -------------------------------
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(2,))
])

model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X,
    y,
    epochs=500,
    verbose=0
)

# -------------------------------
# Loss Curve
# -------------------------------
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'])
plt.title("Training Loss (AND)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# -------------------------------
# Decision Boundary
# -------------------------------
xx, yy = np.meshgrid(
    np.linspace(-0.2,1.2,200),
    np.linspace(-0.2,1.2,200)
)

grid = np.c_[xx.ravel(), yy.ravel()]
pred = model.predict(grid, verbose=0)
pred = pred.reshape(xx.shape)

plt.figure(figsize=(6,6))
plt.contourf(xx, yy, pred, levels=[0,0.5,1], alpha=0.4)

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i,0], X[i,1], color='red', s=100)
    else:
        plt.scatter(X[i,0], X[i,1], color='blue', s=100)

plt.title("Decision Boundary (AND)")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
plt.show()

# -------------------------------
# Predictions
# -------------------------------
predictions = model.predict(X)

print("\nAND Gate Predictions")
print("---------------------")

for i in range(len(X)):
    print(f"Input: {X[i]}  Expected: {int(y[i][0])}  Predicted: {round(predictions[i][0])}")

loss, accuracy = model.evaluate(X, y, verbose=0)

print("\nAccuracy: {:.2f}%".format(accuracy*100))
```
